```python
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://www.propertyfinder.eg/blog/%D8%A3%D9%81%D8%B6%D9%84-%D8%A7%D9%84%D8%A7%D9%85%D8%A7%D9%83%D9%86-%D8%A7%D9%84%D8%B3%D9%8A%D8%A7%D8%AD%D9%8A%D8%A9-%D9%81%D9%8A-%D8%A7%D9%84%D9%82%D8%A7%D9%87%D8%B1%D8%A9/"

# Set headers to simulate a browser visit
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Send request
response = requests.get(url, headers=headers)

# Parse HTML
soup = BeautifulSoup(response.content, "html.parser")

# Get article title
title = soup.find("h1", class_="entry-title").text.strip()
print("Title:", title)

# Get last updated date
date = soup.find("p", class_="post-date").text.strip()
print("Last Updated:", date)

# Find all section headers (h2) that list places
places = soup.find_all("h2")

# Extract place names and their descriptions
for place in places:
    name = place.get_text(strip=True)
    content = []
    for sibling in place.find_next_siblings():
        if sibling.name and sibling.name.startswith("h2"):
            break
        content.append(sibling.get_text(strip=True))
    print("\nPlace:", name)
    print("Details:", " ".join(content[:3]))  # Show first 3 paragraphs
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.propertyfinder.eg/blog/%D8%A3%D9%81%D8%B6%D9%84-%D8%A7%D9%84%D8%A7%D9%85%D8%A7%D9%83%D9%86-%D8%A7%D9%84%D8%B3%D9%8A%D8%A7%D8%AD%D9%8A%D8%A9-%D9%81%D9%8A-%D8%A7%D9%84%D9%82%D8%A7%D9%87%D8%B1%D8%A9/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, 'html.parser')
article = soup.find("article")

sections = []
current_section = None
current_subsection = None

for tag in article.find_all(['h2', 'h3', 'h4', 'p']):
    if tag.name == 'h2':
        current_section = {
            "section_title": tag.get_text(strip=True),
            "subsections": []
        }
        sections.append(current_section)
        current_subsection = None
    elif tag.name == 'h3' and current_section:
        current_subsection = {
            "subheading": tag.get_text(strip=True),
            "paragraphs": [],
            "extra_headings": []
        }
        current_section["subsections"].append(current_subsection)
    elif tag.name == 'p' and current_subsection:
        current_subsection["paragraphs"].append(tag.get_text(strip=True))
    elif tag.name == 'h4' and current_subsection:
        current_subsection["extra_headings"].append(tag.get_text(strip=True))

# Example print
import json
print(json.dumps(sections, ensure_ascii=False, indent=2))
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.propertyfinder.eg/blog/%D8%A3%D9%81%D8%B6%D9%84-%D8%A7%D9%84%D8%A7%D9%85%D8%A7%D9%83%D9%86-%D8%A7%D9%84%D8%B3%D9%8A%D8%A7%D8%AD%D9%8A%D8%A9-%D9%81%D9%8A-%D8%A7%D9%84%D9%82%D8%A7%D9%87%D8%B1%D8%A9/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, 'html.parser')
article = soup.find("article")

sections = []
current_section = None
current_subsection = None

for tag in article.find_all(['h2', 'h3', 'h4', 'p']):
    text = tag.get_text(strip=True)
    if tag.name == 'h2':
        current_section = {"title": text, "subsections": []}
        sections.append(current_section)
        current_subsection = None
    elif tag.name == 'h3' and current_section:
        current_subsection = {"heading": text, "paragraphs": [], "extras": []}
        current_section["subsections"].append(current_subsection)
    elif tag.name == 'p' and current_subsection:
        current_subsection["paragraphs"].append(text)
    elif tag.name == 'h4' and current_subsection:
        current_subsection["extras"].append(text)

# Write to text file
with open("cairo_tourism_guide.txt", "w", encoding="utf-8") as f:
    for section in sections:
        f.write(f"== Section: {section['title']} ==\n\n")
        for sub in section["subsections"]:
            f.write(f"-- {sub['heading']} --\n")
            for para in sub["paragraphs"]:
                f.write(f"{para}\n")
            for extra in sub["extras"]:
                f.write(f"{extra}\n")
            f.write("\n")
```
